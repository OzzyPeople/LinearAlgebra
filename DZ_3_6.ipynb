__1.__ Решить систему уравнений методом Гаусса:

$$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
2x_{1}+x_{2}-x_{3}+x_{4}=-2, \\
x_{1}+x_{2}-3x_{3}+x_{4}=4.
\end{cases}$$

Запишем систему в виде расширенной матрицы:
$${A}=\begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 & -2\\ 
2 & 1 & -1 & 1\\ 
1 & 1 & -3 & 1\\ 
\end{matrix}\right|
\begin{matrix}
0\\ 
-2\\
4
\end{matrix}
\end{pmatrix}$$

Умножим 1-ю строку на (2). Умножим 2-ю строку на (-1). Добавим 2-ю строку к 1-й:

$${A}=\begin{pmatrix}
\left.\begin{matrix}
0 & 1 & -1 & -5\\ 
2 & 1 & -1 & 1\\ 
1 & 1 & -3 & 1\\ 
\end{matrix}\right|
\begin{matrix}
2\\ 
-2\\
4
\end{matrix}
\end{pmatrix}$$


Умножим 3-ю строку на (-2). Добавим 3-ю строку к 2-й:

$${A}=\begin{pmatrix}
\left.\begin{matrix}
0 & 1 & -1 & -5\\ 
0 & -1 & 5 & -1\\ 
1 & 1 & -3 & 1\\ 
\end{matrix}\right|
\begin{matrix}
2\\ 
-10\\
4
\end{matrix}
\end{pmatrix}$$


Добавим 2-ю строку к 1-й:
    
$${A}=\begin{pmatrix}
\left.\begin{matrix}
0 & 0 & 4 & -6\\ 
0 & -1 & 5 & -1\\ 
1 & 1 & -3 & 1\\ 
\end{matrix}\right|
\begin{matrix}
-8\\ 
-10\\
4
\end{matrix}
\end{pmatrix}$$

Теперь исходную систему можно записать так:
$$x3 = (-8 - ( - 6 * x4))/4 \\
x2 = (-10 - (5*x3 - x4))/(-1)\\
x1 = (4 - (x2 - 3*x3 + x4))/1$$

Необходимо переменную x4 принять в качестве свободной переменной и через нее выразить остальные переменные.Приравняем переменную x4 к 0
В итоге, получаем:
$$ x3 = -2 \\
x2 = 0 \\
x1 = -2 $$




__2.__ Проверить на совместность и выяснить, сколько решений будет иметь система линейных уравнений:

   а) $\begin{cases}
3x_{1}-x_{2}+x_{3}=4, \\
2x_{1}-5x_{2}-3x_{3}=-17, \\
x_{1}+x_{2}-x_{3}=0;
\end{cases}$


**Ответ.** Система совместная, так как по методу Гаусса, система имеет 1 решение:

$$ x3 = -80/64 \\
x2 = -2,25 \\
x1 = 1 $$

    
   б) $\begin{cases}
2x_{1}-4x_{2}+6x_{3}=1, \\
x_{1}-2x_{2}+3x_{3}=-2, \\
3x_{1}-6x_{2}+9x_{3}=5;
\end{cases}$
    
**Ответ.** Система не имеет решения, поэтому она несовместная. По методу Гаусса, ранг основной матрицы системы равен r(A)=1. Ранг расширенной матрицы равен r=3 (в основной матрице системы имеется нулевая строка). 

    
   в) $\begin{cases}
x_{1}+2x_{2}+5x_{3}=4, \\
3x_{1}+x_{2}-8x_{3}=-2. 
\end{cases}$

Если количество уравнений меньше, чем количество переменных, то сразу можно сказать, что система либо несовместна, либо имеет бесконечно много решений. 

**Ответ.** Система имеет множество решений, поэтому она неопределенная. По методу Гаусса, когда мы приравниваем свободную переменную x3 к 0,то получаем:
$$
x2 = 2,8 \\
x1 = -1,6 $$

x3 может быть любым значением, поэтому система имеет множество решений.

__4.__ Дана система линейных уравнений, заданная расширенной матрицей

$$\tilde{A}=\begin{pmatrix}
\left.\begin{matrix}
1 & 2 & 3\\ 
4 & 5 & 6\\ 
7 & 8 & 9
\end{matrix}\right|
\begin{matrix}
a\\ 
b\\
c
\end{matrix}
\end{pmatrix}.$$

Найти соотношение между параметрами $a$, $b$ и $c$, при которых система является несовместной.

**Решение:**

Определитель системы равен нулю, поэтому система либо имеет бесконечное множество решений, либо не имеет решений, т.е. несовместна.


1 Составим расширенную матрицу и преобразуем ее к ступенчатой форме:

1.1 Строка 2 - 2* строка 1

$$\tilde{A}=\begin{pmatrix}
\left.\begin{matrix}
1 & 2 & 3\\ 
2 & 1 & 0\\ 
7 & 8 & 9
\end{matrix}\right|
\begin{matrix}
a\\ 
b-2a\\
c
\end{matrix}
\end{pmatrix}.$$

1.2 Строка 3 - 3* строка 1

$$\tilde{A}=\begin{pmatrix}
\left.\begin{matrix}
1 & 2 & 3\\ 
2 & 1 & 0\\ 
4 & 2 & 0
\end{matrix}\right|
\begin{matrix}
a\\ 
b-2a\\
c-3a
\end{matrix}
\end{pmatrix}.$$

1.2 Строка 3 - 2* строка 2

$$\tilde{A}=\begin{pmatrix}
\left.\begin{matrix}
1 & 2 & 3\\ 
2 & 1 & 0\\ 
0 & 0 & 0
\end{matrix}\right|
\begin{matrix}
a\\ 
b-2a\\
c-3a-2b
\end{matrix}
\end{pmatrix}.$$

**Ответ:** Если c-3a-2b не равно 0, то система является несовместной. В противном случае одна из неизвестных является свободной переменной и, следовательно, система имеет бесконечное множество решений.
