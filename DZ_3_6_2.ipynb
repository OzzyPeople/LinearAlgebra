   а) $\begin{cases}

$$ \Delta = 1*(-4) - (-2)*3 = -4 + 6 = 2 \neq 0 $$

$$ \Delta{x1} = 1*(-4) - (-2)*7 = -4 + 14 = 10 $$
$$ \Delta{x2} = 1*(7) - (1)*3 = 7 - 3 = 4 $$

$$ x1 = \Delta{x1} / \Delta = 10/2 = 5$$

$$ x2 = \Delta{x2} / \Delta = 4 /2 = 2$$

__4.__ Решить систему линейных уравнений методом Холецкого

$$\begin{cases}

-45x_{1}+50x_{2}-15x_{3}=-460 \\
45x_{1}-15x_{2}+38x_{3}=193
\end{cases}$$

Пусть 
$$A =\begin{pmatrix}
81 & -45 & 45 \\ 
-45 & 50 & -15 \\ 
45 & -15 & 38
\end{pmatrix}
b =\begin{pmatrix}
531 \\ 
-460 \\ 
193
\end{pmatrix}$$

Находим элементы матрицы L, для этого произведем разложение на $LL^{T}$:

$$l_{11}=\sqrt{a_{11}}=\sqrt{81} = 9,$$
$$l_{21}=\frac{a_{21}}{l_{11}}=\frac{-45}{9} = -5,$$
$$l_{31}=\frac{a_{31}}{l_{11}}=\frac{45}{9} = 5,$$

$$l_{22}=\sqrt{a_{22}-l_{21}^{2}}=\sqrt{50 - (-5)^2}=5,$$
$$l_{32}=\frac{1}{l_{22}}\left ( a_{32}-l_{21}l_{31} \right) = \frac {( -15 - (-5)*5)}{5} = 2$$

$$l_{33}=\sqrt{a_{33}-l_{31}^{2}-l_{32}^{2}}=\sqrt{38-5^2-2^2}=3 $$


Таким образом разложение матрицы A имеет вид:

$$A =LL^t=\begin{pmatrix}
9 & 0 & 0 \\ 
-5 & 5 & 0 \\ 
5 & -2 & 3
\end{pmatrix}
\begin{pmatrix}
9 & -5 & 5 \\ 
0 & 5 & -2 \\ 
0 & -0 & 3
\end{pmatrix}$$
